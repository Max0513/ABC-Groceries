# Association Rule Leearning (Apriori)

```python
# Import Packages
from apyori import apriori
import pandas as pd
```

```python
# Import Data
alcohol_transactions = pd.read_csv('../../Data/sample_data_apriori.csv')
```

```python
alcohol_transactions.head()
```

```python
alcohol_transactions.shape
```

```python
# Drop ID

alcohol_transactions.drop(['transaction_id'], axis=1,inplace=True)
```

```python
# Create a list of transactions

transactions_list = []


for index, row in alcohol_transactions.iterrows():
    transactions = list(row.dropna())
    transactions_list.append(transactions)

transactions_list[:2]
```

```python
# Apply Apriori

apriori_rules = apriori(transactions_list,
                        min_support=0.003,  # Minimum support to be considered by algorythm
                        min_confidence=0.2, # Minimum confidence to be considered by algorythm
                        min_lift=3, # Minimum lift to be considered by algorythm
                        min_length=2, # Min number of items to be compared
                        max_length=2 # Max number of items to be compared
                        )
apriori_rules = list(apriori_rules)
```

Support is the percentage of transactions that contain all items 

confidence is the percentage of transactions that contain both transactions A and B out of transactions that contain item A 

lift is the ration of confidence over the expected confidence

```python
# Convert back to dataframe

# Get objects for DataFrame
product1 = [list(rule[2][0][0])[0] for rule in apriori_rules]
product2 = [list(rule[2][0][1])[0] for rule in apriori_rules]
support = [rule[1] for rule in apriori_rules]
confidence = [rule[2][0][2] for rule in apriori_rules]
lift = [rule[2][0][3] for rule in apriori_rules]

# Create DataFrame
apriori_df = pd.DataFrame({'Product 1':product1,
                           'Product 2':product2,
                           'Support':support,
                           'Confidence':confidence,
                           'Lift':lift
                           }
                          )
apriori_df.sort_values('Lift', ascending=False, inplace=True)
```
